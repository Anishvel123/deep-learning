```python
import numpy as np
import tensorflow_datasets as tfds
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
dataset, info = tfds.load('imdb_reviews', with_info = True, as_supervised = True)
```

```python
train_sentences = []
train_labels = []
```

```python
train_data = dataset['train']
for s, l in train_data:
  train_sentences.append(s.numpy().decode('utf-8'))
  train_labels.append(l.numpy())
```

```python
import json
import pickle
with open('train_data.json', 'w', encoding = 'utf-8') as f:
  json.dump(train_sentences, f, ensure_ascii=False)
with open('train_labels.pkl', 'wb') as f:
  pickle.dump(train_labels, f)
```

```python
import json
import pickle
with open('train_data.json', 'r', encoding = 'utf-8') as f:
   train_sentences= json.load(f)
with open('train_labels.pkl', 'rb') as f:
  train_labels = pickle.load(f)
```

```python
vocab_size = 10000
embedded_vector_size = 64
max_len = 200
```

```python
encoded_reviews = [one_hot(rev, vocab_size) for rev in train_sentences]
padded_reviews = pad_sequences(encoded_reviews, maxlen=max_len, padding='post')
```

```python
model = Sequential()
model.add(Embedding(vocab_size, embedded_vector_size, input_length = max_len, name = 'embeddings'))
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
X = padded_reviews
y = np.array(train_labels)
```

```python
model.fit(X, y, epochs=20)
```

```python
word_embeddings = np.array(model.get_layer('embeddings').get_weights()[0])
```

```python
with open('word_embeddings.pkl', 'wb') as f:
  pickle.dump(word_embeddings, f)
```
